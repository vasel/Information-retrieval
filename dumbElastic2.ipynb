```python
def normaliza(termo):
    return stemmer.stem(unidecode.unidecode(termo.lower()).replace(",","").replace('.',''))
```

```python
dados = {}
def insereDados(arquivo): 
    f = open(arquivo,'r', encoding="utf8") 
    dados[arquivo] = f.read()
```
